```python
plt.figure(figsize=(20,6))
plt.xticks(rotation=90)
sns.barplot(data=df, x='Region', y='Estimated Unemployment Rate (%)',palette = 'Set1', order=df.groupby('Region')['Estimated Employed'].mean().sort_values().index)
plt.title('Estimated Unemployment Rate (%) by Region')
plt.show()
```

Tripura has the highest unemployment rate.

```python
plt.figure(figsize=(20,6))
plt.xticks(rotation=90)
sns.barplot(data=df, x='Region', y='Estimated Employed',palette = 'Set1', order=df.groupby('Region')['Estimated Employed'].mean().sort_values().index)
plt.title('Estimated Employed by Region')
plt.show()
```

Uttar Pradesh has the most employments.

```python
plt.figure(figsize=(20,6))
plt.xticks(rotation=90)
sns.barplot(data=df, x='Region', y='Estimated Labour Participation Rate (%)',palette = 'Set1', order=df.groupby('Region')['Estimated Employed'].mean().sort_values().index)
plt.title('Estimated Labour Participation Rate (%) by Region')
plt.show()
```

Tripura has the highest Labour Participation Rate (%)

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.barplot(data=df, x='Area', y='Estimated Labour Participation Rate (%)',palette = 'Set1', order=df.groupby('Area')['Estimated Employed'].mean().sort_values().index)
plt.title('Estimated Labour Participation Rate (%) by Area')
plt.show()
```

Rural areas have a higher labour participation rate.


```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.barplot(data=df, x='Area', y='Estimated Employed',palette = 'Set1', order=df.groupby('Area')['Estimated Employed'].mean().sort_values().index)
plt.title('Estimated Employed by Area')
plt.show()
```

Rural areas have a higher employment.

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.barplot(data=df, x='Area', y='Estimated Unemployment Rate (%)',palette = 'Set1', order=df.groupby('Area')['Estimated Employed'].mean().sort_values().index)
plt.title('Estimated Unemployment Rate (%) by Area')
plt.show()
```

Urban areas have a higher unemployment rate.

```python
unemployment_df = df[['Region', 'Area', 'Estimated Unemployment Rate (%)', 'Estimated Employed', 'Estimated Labour Participation Rate (%)']]
unemployment = unemployment_df.groupby(['Region', 'Area'])['Estimated Unemployment Rate (%)'].mean().reset_index()
fig = px.sunburst(unemployment, path=['Region', 'Area'], values='Estimated Unemployment Rate (%)',
                  color_continuous_scale='Plasma', title='Unemployment rate in each state',
                  height=650, template='ggplot2')
fig.show()
```

The sunburst chart shows the regions in india and their unemployment rates in both rural and urban countries.
